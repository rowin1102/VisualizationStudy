```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터프레임으로 변환시 header 옵션이 없으므로 첫 행은 컬럼면으로 인식
df = pd.read_excel('../resData/남북한_발전_전력량.xlsx', engine='openpyxl')
```

```python
df
```

```python
# [0, 5] -> 엑셀의 0행과 5행을 의미
# 3: -> D열부터 마지막을 의미
df_ns = df.iloc[[0, 5], 3:]
# 남한과 북한의 합계 부분만 추출
df_ns
```

```python
# 남북한 데이터에 인덱스 부여
df_ns.index = ['South', 'North']
df_ns
```

```python
# 열 이름의 자료형을 정수형으로 변경
df_ns.columns = df_ns.columns.map(int)
df_ns.columns
```

```python
print(f"{'원본 데이터':-^30}")
print(df_ns.head())
print(f"{'전치한 데이터':-^30}")
print(df_ns.T.head())
```

```python
# 선 그래프1(원본 데이터프레임)
df_ns.plot()
plt.show()
```

```python
# 선 그래프2(전치한 데이터프레임)
df_ns.T.plot()
plt.show()
```

```python
# 막대 그래프1(원본 데이터프레임)
df_ns.plot(kind='bar')
plt.show()
```

```python
# 막대 그래프2(전치한 데이터프레임)
df_ns.T.plot(kind='bar')
plt.show()
```

```python
# 히스토그램(전치한 데이터프레임)
df_ns = df_ns.apply(pd.to_numeric, errors='coerce')
df_ns.T.plot(kind='hist')
plt.show()
```
